```python
from apify_client import ApifyClient
import pandas as pd
import os

API_TOKEN1 = os.getenv("APIFY_TOKEN1")
API_TOKEN2 = os.getenv("APIFY_TOKEN2")

client1 = ApifyClient("API_TOKEN1")
client2 = ApifyClient("API_TOKEN2")


run_input1 = {
    "language": "en",
    "maxReviews": 11000,
    "personalData": True,
    "reviewsSort": "newest",
    "startUrls": [
        {
            "url": "https://maps.app.goo.gl/da9XMKWSN6eGLRka8"
        }
    ]
}

run_input2 = {
    "language": "en",
    "maxReviews": 11000,
    "personalData": True,
    "reviewsSort": "mostRelevant",
    "startUrls": [
        {
            "url": "https://maps.app.goo.gl/da9XMKWSN6eGLRka8"
        }
    ]
}

run1 = client1.actor("Xb8osYTtOjlsgI6k9").call(run_input=run_input1)
run2 = client2.actor("Xb8osYTtOjlsgI6k9").call(run_input=run_input2)

result1 = []
for item in client1.dataset(run1["defaultDatasetId"]).iterate_items():
    result1.append(item)

result2 = []
for item in client2.dataset(run2["defaultDatasetId"]).iterate_items():
    result2.append(item)
```

```python
df1 = pd.DataFrame(result1)
df2 = pd.DataFrame(result2)

df = pd.concat([df1, df2], ignore_index=True)

df.to_csv("scraping_pjv_last.csv", index=False, encoding='utf-8')
```
